```python
# Type Casting
# is a process of converting one data type into another data type

# intValue ---> string
# str(intValue)

# floatValue ---> string
# str(floatValue)

# intValue --> float
# float(intValue)

# floatValue --> int
# int(floatValue)

# boolValue --> string
# str(boolValue)

# integer value --> string
age = 19
print(age, type(age))

age = str(age)
print(age, type(age))

# float value --> int
result = 90.000445642
print(result, type(result))

result = int(result)
print(result, type(result))
```

```python
# Operators
# Are used to perform operations on variables

# Arithmetic Operators + - * / %  // **

a = 90
b = 5

print(a+b)
print(a-b)
print(a*b)
print(a/b)    # Division Operator return value in FLOAT
print(a//b)   # Division Operator return value in INTEGER
print(a%b)    # gives a remainder (baki dari bahagi)
print(a**b)   # a raise to the power of b [90 raise to the power of 5]
```

```python
# Logical operator
# AND OR NOT

# AND operator requires more than one condition
# AND operator returns TRUE only when both the conditions are TRUE

#print(100==100 and 50>100)

nat = input('Nationality : ')
age = int(input('Age : '))

print(nat=="Malaysian" or nat=="Singaporean" and age>18)
```

```python
# OR operator required more than one condition
# OR operator returs TRUE when both conditions are TRUE
# OR operator returs FALSE when both conditions are FALSE
# OR operator returs TRUE when either one of conditions are TRUE


# OR operator requires more than one condition
# OR operator returns TRUE when both the conditions are TRUE
# OR operator returns TRUE if any one of the conditions is TRUE
# OR operator returns FALSE when both the conditions are FALSE

print(100==100 or 50<100)
#      cond1        cond2
#       true         true


print(100==100 or 50>100)
#      cond1        cond2
#       true        false

print(10==100 or 50>100)
#      cond1        cond2
#      false        False

# NOT operator
# inverse the result
# True --> False     False --> True

print(not(100==100 and 50<100))
```

```python
# Membership Operator -- used for searching operations
# in not in

name = "John Smith"
print('o' in name)
print('Smith'in name)

# not in
print('J' not in name)  # 'J' does not exist in name
print('s' not in name)  # 's' does not exist in name
```
