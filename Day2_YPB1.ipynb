```python
# Data Types
# data type defines what kind of a value a variable can hold

# Python is a Dynamically Typed Programming Language
#   a.) we don't need to write data type before declaring a variable
#   b.) we can re-assign diff diff values to a variable

# INT : used to hold a numeric value [positive / negative]

a = 18
print(a)

# type(varname) : used to determine the data type of a variable

print(type(a))

b = 'Python'
print(b, type(b))

c = True
print(c,type(c))

d = 20.1
print (d,type(d))
```

```python
# Indexing
# indexing allows us to retrieve only one single value
# indexing is denoted by using [] square brackets

#  W   I   R   A
#  0   1   2   3      forward / positive index numbers
# -4  -3  -2  -1      backward / negative index numbers

str2 = "WIRA"
print (str2[3])
print (str2[-4])
```

```python
# slicing allows to retrieve portion / sub-portion of a sequence
# syntax :   varname[startIndex:stopIndex]
# startIndex : index no. from where slicing will begin     INCLUSIVE
# stopIndex  : index no. where slicing will end            EXCLUSIVE

course = 'ESS_959027_MBS_IR_Podium_L1'
print(course[0:9])
print(course[9:13])
print(course[4:9])
print(course[6:17])

# in slicing if stopIndex is missing then slicing will go through till the last character

print(course[6:])

# in slicing if startIndex is missing then slicing will begin from index ZERO

print(course[:10])



# syntax : varname[startIndex:stopIndex:stepCount/interval]

print(course[3:30:2])
```

```python
# String Concatination / Formatting

name = "John"
course = "Python"
fees = 1500

# John is learning Python and the fees is RM1500

# + operator requires left hand and right hand side value to be of type STRING
# use str(var) to convert int to string - type casting
print(name+" is learning "+course+ ' and the fee is RM'+str(fees)+'')
print(name,"is learning",course,'and the fee is RM'+str(fees))

# Formatting - variables are degined in curly braces
print(f'{name} is learning {course} and the fee is RM{fees}')
print(f'{name} is learning {course} and the fee is RM{fees+210}')
```
